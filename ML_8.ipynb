1.What is unsupervised learning in the context of machine learning?

Ans :- Unsupervised learning is a type of machine learning in which the model learns patterns, structures, or relationships from unlabeled data. The goal is to discover hidden patterns, group similar data points, or reduce data dimensions without predefined target values.

2. How does the K-Means clustering algorithm work?

Ans :- K-Means clustering works by dividing data into K clusters. It starts by randomly selecting K centroids, assigns each data point to the nearest centroid, recalculates the centroids based on the assigned points, and repeats the process until the centroids no longer change significantly.

3. Explain the concept of a dendrogram in hierarchical clustering.

Ans :- A dendrogram is a tree-like diagram used in hierarchical clustering to visualize how clusters are formed. It shows the sequence of merges or splits and helps determine the optimal number of clusters by observing where large distances occur between cluster combinations.

4. What is the main difference between K-Means and Hierarchical Clustering?

Ans :- K-Means requires the number of clusters to be specified beforehand and partitions data into fixed clusters, whereas Hierarchical Clustering creates a hierarchy of clusters without requiring a predefined number of clusters.

5.What are the advantages of DBSCAN over K-Means?

Ans :- DBSCAN can identify clusters of arbitrary shapes, automatically detect outliers or noise points, and does not require specifying the number of clusters in advance. It also performs better when clusters have irregular shapes.

6. When would you use Silhouette Score in clustering?
Silhouette Score is used to evaluate the quality of clustering by measuring how similar a data point is to its own cluster compared to other clusters. It is commonly used to determine the optimal number of clusters.

7. What are the limitations of Hierarchical Clustering?

Ans :- Hierarchical Clustering is computationally expensive for large datasets, sensitive to noise and outliers, and once clusters are merged or split, the process cannot be reversed.

8. Why is feature scaling important in clustering algorithms like K-Means?

Ans :- Feature scaling ensures that all features contribute equally to distance calculations. Without scaling, features with larger numerical ranges can dominate the clustering process and produce biased results.

9. How does DBSCAN identify noise points?

Ans :- DBSCAN identifies noise points as those that do not belong to any dense region. Specifically, points that have fewer than the minimum required neighboring points within a specified radius are labeled as noise.

10.define inertia in the context of K-Means.

Ans :- Inertia is the sum of squared distances between each data point and the centroid of its assigned cluster. It measures cluster compactness, with lower inertia indicating tighter clusters.

11.What is the elbow method in K-Means clustering?

Ans :- The elbow method helps determine the optimal number of clusters by plotting inertia against the number of clusters. The point where the rate of decrease sharply changes, forming an "elbow," is considered the optimal number of clusters.

12. Describe the concept of "density" in DBSCAN.

Ans :- In DBSCAN, density refers to the concentration of data points within a specified neighborhood radius. Regions with sufficient nearby points are considered dense and form clusters.

13. Can hierarchical clustering be used on categorical data?

Ans :- Yes, hierarchical clustering can be applied to categorical data by using appropriate distance or similarity measures, such as Hamming distance or other categorical similarity metrics.

14. What does a negative Silhouette Score indicate?

Ans :- A negative Silhouette Score indicates that a data point is likely assigned to the wrong cluster because it is closer to points in another cluster than to points in its own cluster.

15.Explain the term "linkage criteria" in hierarchical clustering.

Ans :- Linkage criteria determine how the distance between clusters is calculated during hierarchical clustering. Common methods include single linkage, complete linkage, average linkage, and Ward's linkage.

16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans :- K-Means assumes clusters are spherical, equally sized, and have similar densities. When clusters vary significantly in size or density, K-Means may incorrectly assign data points and produce poor clustering results.

17. What are the core parameters in DBSCAN, and how do they influence clustering?

Ans :- The two main parameters are epsilon (ε) and MinPts. Epsilon defines the neighborhood radius, while MinPts specifies the minimum number of points required to form a dense region. These parameters determine cluster formation and noise detection.

18. How does K-Means++ improve upon standard K-Means initialization?

Ans :- K-Means++ selects initial centroids more strategically by spreading them apart, which reduces the chances of poor initialization, improves clustering accuracy, and often speeds up convergence.

19. What is agglomerative clustering?

Ans :- Agglomerative clustering is a bottom-up hierarchical clustering approach where each data point starts as its own cluster, and the closest clusters are repeatedly merged until a stopping criterion is reached.

20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Ans :- Silhouette Score evaluates both cluster cohesion (how close points are within a cluster) and cluster separation (how distinct clusters are from each other). In contrast, inertia only measures cluster compactness and always decreases as the number of clusters increases, making it less reliable for determining the optimal clustering structure.
